        html = requests.get(link).content
        soup_anos = BeautifulSoup(html, 'html.parser')
        for citação in soup_anos.find_all('div', 'gs_a'):
            print('Esperando 45 segundos.')
            time.sleep(45)
            ano = int(citação.text.split(' - ')[-2].split(', ')[-1])
            anos.append(ano)

        sopa2 = soup_anos.find(id="gs_n")
        if not sopa2 == None:
            url_parcial = sopa2.center.table.tr.a['href']
            href = r'https://scholar.google.com.br' + sopa2
            anos = get_ano_das_citações (href, anos)

        return anos
