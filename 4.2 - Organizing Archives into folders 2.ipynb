```python
old_folder = r'\\tsclient\E\Lattes\Lattes_ZIP'
new_folder = r'D:\Lattes\Lattes_ZIP'
sobrescreve = False
```

## Creating SubFolders

```python
# Creating Subdirectorys, if they don't exixsts

if not os.path.isdir(new_folder):
    os.makedirs(new_folder)
    #print("created folder : ", new_folder)

for x in range (10):
    file_path1 = os.path.join(new_folder, str(x))
    for y in range(10):
        file_path2 = os.path.join(file_path1, str(y))
        CHECK_FOLDER = os.path.isdir(file_path2)
        #print(file_path2, CHECK_FOLDER)
                

        # If folder doesn't exist, then create it.
        if not CHECK_FOLDER:
            os.makedirs(file_path2)
            #print("created folder : ", file_path2)

        else:
            #print(file_path2, "folder already exists.")
            pass
```

# Arrumando arquivos j√° existentes

```python
old_files = [y for x in os.walk(old_folder) for y in glob.glob(os.path.join(x[0], '*.zip'))]
```

```python
new_files = [y for x in os.walk(new_folder) for y in glob.glob(os.path.join(x[0], '*.zip'))]
```

```python
len(new_files)
```

```python
#Verificando arquivos j√° existente
lista_ids_j√°_existentes = []
arquivos_no_lugar_errado = []
if not sobrescreve:
    total_files = len(old_files)
    files_moved = 0
    num_files = -1
    for f in new_files:
        if files_moved % 1000 == 0:
            print(f'{files_moved}/{total_files}: {f}')
        if f.endswith('.zip'):
            lattes_id = re.findall(r'\d+', os.path.basename(f))[0]
            while len(lattes_id) < 16:
                lattes_id = '0' + lattes_id
            lista_ids_j√°_existentes.append(lattes_id)
            base_path = os.path.dirname(f)
            new_path = os.path.dirname(os.path.join(new_folder, lattes_id[0], lattes_id[1], 'Lattes_' + Path(lattes_id).stem+ '.zip'))
            if not base_path == new_path:
                if files_moved % 1000 == 0: print(f'{files_moved}/{total_files}: error in file {f}')
                arquivos_no_lugar_errado.append(f)
            files_moved += 1
    print('all files moved')    
```

```python
#files = os.listdir(old_folder)
list_files = []
n√∫mero_total = len(old_files)
files_moved = 0
files_skipped = 0
num_files = -1
lista_regress√µes = []
tempo_in√≠cio = datetime.now()
feitos = 0

for f in old_files:
    
    #Realizando previs√£o de t√©rmino
    if not feitos == 0:
        porcentagem_j√°_feita = (feitos/n√∫mero_total)
        tempo_passado = datetime.now() - tempo_in√≠cio
        tempo_por_id = tempo_passado / feitos
        tempo_restante = (n√∫mero_total - feitos) * tempo_por_id
        tempo_em_que_vai_acabar = datetime.now() + tempo_restante
        if feitos % 1000 == 0:
            print(f'{feitos}/{n√∫mero_total}. {porcentagem_j√°_feita * 100}% feitos. Fazendo: {f}. Acabar√° em {tempo_em_que_vai_acabar}. O √öltimo demorou: {datetime.now() - tempo_√∫ltimo}.')
    else:
        if feitos % 1000 == 0:
            print(f'{feitos}/{n√∫mero_total}. Fazendo: {f}.')
    feitos += 1
    tempo_√∫ltimo = datetime.now() 
    
    if f.endswith('.zip'):
        files_moved += 1
        lattes_id = re.findall(r'\d+', os.path.basename(f))[0]
        while len(lattes_id) < 16:
            lattes_id = '0' + lattes_id
#        if not sobrescreve:
#            if lattes_id in lista_ids_j√°_existentes:
#                lista_ids_j√°_existentes.remove(lattes_id)
#                files_skipped += 1
#                continue
        oldfile = os.path.join(old_folder, f) 
        newfile = os.path.join(new_folder, lattes_id[0], lattes_id[1], 'Lattes_' + Path(lattes_id).stem+ '.zip')
        if sobrescreve or not os.path.isfile(newfile):
            try:
                shutil.copyfile(oldfile, newfile)
                list_files.append({
                    'sucess': True,
                    'Old:': oldfile,
                    'New:': newfile
                })
            except:
                list_files.append({
                    'sucess': False,
                    'Old:': oldfile,
                    'New:': newfile
                })
                
        
        if num_files >=0 and files_moved > num_files:
              break

print('all files moved')
```
