```python
from io import BytesIO, StringIO
from zipfile import ZipFile
import zeep
import xmltodict, json

import io, sys, csv, html, os
```

```python
settings = zeep.Settings(strict=True, xml_huge_tree=True)           
client = zeep.Client(wsdl='http://servicosweb.cnpq.br/srvcurriculo/WSCurriculo?wsdl')
lattes_dt_atualizacao = client.service.getDataAtualizacaoCV('7281587998425548')
```

```python
print(lattes_dt_atualizacao)
```

```python
zip_lattes = client.service.getCurriculoCompactado('7281587998425548')
```

```python
with ZipFile(io.BytesIO(zip_lattes)) as myzip:
        with myzip.open(myzip.namelist()[0]) as myfile:
                    lattes = myfile.read()

```

```python
print(lattes)
```

```python
filename = 'd:\Lattes_7281587998425548.zip'
with open(filename , 'wb') as f:
    f.write(zip_lattes)
```
